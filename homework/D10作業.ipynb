### 1. [簡答題] 請問下列四種不同的 DataFrame 選取結果有什麼差異？

```
df.loc[ '2013-01-01', 'A'] 
df.loc[ '2013-01-01', ['A', 'B'] ] 
df.loc[ '2013-01-01':'2013-01-02', 'A' ] 
df.loc[ '2013-01-01':'2013-01-05', 'A':'C'] 
```





### answer
1. df.loc[ '2013-01-01', 'A']: 會選取到一個值，此做法為一個座標圖
2. df.loc[ '2013-01-01', ['A', 'B'] ]: 會得到列位名稱"2013-01-01"的"A","B"欄位的值
3. df.loc[ '2013-01-01':'2013-01-02'. 'A']: 會得到"A"欄位，列位"2013-01-01"到'2013-01-02"的值
4. df.loc[ '2013-01-01':'2013-01-05'. 'A':'C']: 會得到"A""B""C"欄位的"01"到"05"的值

### 2. 請根據提供的資料，選擇出下列的要求：

```
# - select the first 3 rows.
# - select the odd rows. (index = 1, 3, 5)
# - select the last 2 columns.
# - select the even columns. (index = 0, 2, 4)
```


````python
df = pd.DataFrame(np.random.randn(6,3),columns=['col1','col2','col3'])
df
````

````python
'''
Your Code
'''
print(df[0:3]) # select the first 3 rows.
````

````python
df.iloc[lambda x: x.index % 2 == 1] # select the odd rows.
````

````python
df.iloc[:,[-2,-1]] #select the last 2 columns.
````

````python
df.iloc[:,[0,2]] # select the even columns.
````

### 3. 請根據提供的資料，選擇出下列的要求：

```
# - 1. filtered by first column > 20?
# - 2. filtered by first column + second column > 50
# - 3. filtered by first column < 30 or second column > 30
# - 4. filtered by total sum of row > 100
```

````python
# df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4), columns = ['A','B','C','D'])
df
````

````python
'''
Your Code
'''
df.loc[df['A'] > 20,['A']] # filtered by first column > 20?
````

````python
df['A+B']= df['A'] + df['B']
df.loc[df['A+B']>50, ['A+B']] # filtered by first column + second column > 50
````

````python
del df['A+B']
````

````python
df.loc[(df['A'] < 30)|(df['B'] > 30),['A','B']] # filtered by first column < 30 or second column > 30
````

````python
df['sum_row']= df.A + df.B + df.C + df.D
df.loc[df['sum_row'] > 100, ['sum_row']] # filtered by total sum of row > 100
````

````python
del df['sum_row']
df
````
